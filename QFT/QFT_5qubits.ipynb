```python
from qiskit import QuantumCircuit
import numpy as np
```

```python
def qft_using5qubits():
  qc = QuantumCircuit(5)
  #Block 1
  qc.h(0);
  qc.cp(np.pi/2,1,0);
  qc.cp(np.pi/4,2,0);
  qc.cp(np.pi/8,3,0);
  qc.cp(np.pi/16,4,0);

  #Block 2
  qc.h(1);
  qc.cp(np.pi/2,2,1);
  qc.cp(np.pi/4,3,1);
  qc.cp(np.pi/8,4,1);

  #Block 3
  qc.h(2);
  qc.cp(np.pi/2,3,2);
  qc.cp(np.pi/4,4,2);

  #Block 4
  qc.h(3);
  qc.cp(np.pi/2,4,3);

  #Swap gates
  qc.swap(0,4);
  qc.swap(1,3);
  return qc
```

```python
qc = qft_using5qubits()
qc.draw('mpl')
```
