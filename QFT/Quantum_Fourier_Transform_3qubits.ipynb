```python
from qiskit import QuantumCircuit
import numpy as np
from qiskit.visualization import plot_histogram
```

```python
def qft_using3qubits():
  qc = QuantumCircuit(3)
  qc.h(0)
  qc.cp(np.pi/2,1,0)
  qc.cp(np.pi/4,2,0)
  #First block completed
  qc.h(1)
  qc.cp(np.pi/2,2,1)
  #Second block completed
  qc.h(2)
  qc.swap(0,2)
  return qc
```

```python
qc = qft_using3qubits()
qc.draw('mpl')
```
